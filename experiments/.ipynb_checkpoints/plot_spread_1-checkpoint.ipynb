```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Plot Bid/Ask Spread

```python
with open('snapshot_history.pkl', 'rb') as f:
    snapshot_history = pickle.load(f)
history = snapshot_history['spreads']

#
df = pd.DataFrame(columns=['time', 'bid', 'ask'])
df['time'] = [x[0] for x in history]
df['bid'] = [x[1][0] if x[1] is not None else np.nan for x in history]
df['ask'] = [x[1][1] if x[1] is not None else np.nan for x in history]

df['price'] = (df['ask'] + df['bid']) / 2.
df['spread'] = df['ask'] - df['bid']
df['rolling'] = df['spread'].rolling(15, win_type ='triang').mean() 
# df.head().T
```

```python
#
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,8))
ax2 = ax[1].twinx()

ax[0].plot(df['time'], df['ask'], color='red', linestyle='-')
ax[0].plot(df['time'], df['bid'], color='green', linestyle='-')
ax[0].set_ylabel('BID/ASK')

# Real price (?)
# df['real'] = 30*(1 - (df['time']/30)**1.5 + (df['time']/40)**2.5)
# ax[0].plot(df['time'], df['real'], color='blue', linestyle='-')

ax[1].set_ylabel('Spread')
ax[1].plot(df['time'], df['rolling'], color='blue', linestyle='-')

ax2.plot(df['time'], df['price'], color='black', linestyle='-')
ax[1].set_ylim(0,None)
ax2.set_ylim(0,None)
ax2.set_ylabel('Price')
# ax2.set_ylim(-20,20)
```
