```python
from matplotlib import pyplot as plt
import random
import numpy as np

def expo(param):
    x = [random.expovariate(param) for r in range(100)]                   
    plt.hist(x,10)
    plt.show()
    
def normal(mu, sigma):
    x = [random.normalvariate(mu, sigma) for r in range(100)]                   
    plt.hist(x,10)
    plt.show()
```

```python
normal(50,1)
```
